```python
for x, y in train_dataloader:  # 학습 데이터 배치 루프
    x = x.to(device)  # 데이터를 장치로 이동
    y = y.to(device)  # 레이블을 장치로 이동

    output = model(x)  # 모델 예측
    loss = criterion(output, y)  # 손실 계산
    
    optimizer.zero_grad()  # 기존 기울기 초기화
    loss.backward()  # 역전파를 통한 기울기 계산

    torch.nn.utils.clip_grad_norm_(model.parameters(), 0.1)  # 기울기 폭주를 막기 위해 기울기의 노름(norm)을 최대 0.1로 제한(클리핑)합니다.

    optimizer.step()  # 가중치 업데이트
```
