```python
from torch import nn  # 신경망 모듈을 임포트합니다.


class Net(nn.Module):  # nn.Module을 상속받아 신경망 클래스를 정의합니다.
    def __init__(self):  # 생성자 함수 정의
        super().__init__()  # 부모 클래스의 생성자를 호출합니다.
        self.layer = nn.Sequential(  # 여러 층을 순차적으로 쌓는 Sequential 컨테이너를 생성합니다.
            nn.Linear(1, 2),  # 입력 1개, 출력 2개인 선형 층(Linear Layer)을 추가합니다.
            nn.Sigmoid()      # 활성화 함수로 시그모이드(Sigmoid)를 추가합니다.
        )
        self.fc = nn.Linear(2, 1)  # 입력 2개, 출력 1개인 마지막 완전 연결 층(Fully Connected Layer)을 정의합니다.
        self._init_weights()  # 가중치 초기화 함수를 호출합니다.

    def _init_weights(self):  # 가중치를 수동으로 초기화하는 메서드입니다.
        nn.init.xavier_uniform_(self.layer[0].weight)  # 첫 번째 층의 가중치를 Xavier Uniform 방식으로 초기화합니다.
        self.layer[0].bias.data.fill_(0.01)  # 첫 번째 층의 편향(bias)을 0.01로 채웁니다.

        nn.init.xavier_uniform_(self.fc.weight)  # 마지막 층의 가중치를 Xavier Uniform 방식으로 초기화합니다.
        self.fc.bias.data.fill_(0.01)  # 마지막 층의 편향을 0.01로 채웁니다.


model = Net()  # 정의한 신경망 모델의 인스턴스를 생성합니다.
```
