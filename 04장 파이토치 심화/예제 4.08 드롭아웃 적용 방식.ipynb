```python
from torch import nn  # 신경망 모듈을 임포트합니다.


class Net(nn.Module):  # 신경망 클래스 구성
    def __init__(self):  # 초기화 메서드
        super().__init__()  # 부모 클래스 초기화
        self.layer1 = nn.Linear(10, 10)  # 10개의 입력을 받아 10개를 출력하는 첫 번째 선형 층
        self.dropout = nn.Dropout(p=0.5)  # 50%의 확률로 뉴런을 비활성화하는 드롭아웃 층을 정의합니다.
        self.layer2 = nn.Linear(10, 10)  # 10개의 입력을 받아 10개를 출력하는 두 번째 선형 층

    def forward(self, x):  # 순전파 연산 정의
        x = self.layer1(x)  # 첫 번째 층 통과
        x = self.dropout(x)  # 드롭아웃 적용 (학습 시에만 무작위로 뉴런을 끕니다)
        x = self.layer2(x)  # 두 번째 층 통과
        return x
```
