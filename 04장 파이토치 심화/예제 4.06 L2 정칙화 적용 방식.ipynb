```python
for x, y in train_dataloader:  # 학습 데이터로더에서 배치 데이터를 가져오는 루프입니다.
    x = x.to(device)  # 데이터를 연산 장치(GPU 등)로 이동시킵니다.
    y = y.to(device)  # 레이블을 연산 장치로 이동시킵니다.

    output = model(x)  # 모델을 통해 예측 결과(출력)를 계산합니다.

    _lambda = 0.5  # L2 정칙화 가중치를 결정하는 람다값을 설정합니다.
    l2_loss = sum(p.pow(2.0).sum() for p in model.parameters())  # 모든 파라미터의 제곱 합을 구합니다 (L2 손실).

    loss = criterion(output, y) + _lambda * l2_loss  # 오차에 L2 손실을 가산하여 과적합을 방지합니다.
```
