```python
import nlpaug.augmenter.word as naw  # 단어 단위 증강 도구를 임포트합니다.


texts = [  # 증강할 원본 텍스트 리스트
    "Those who can imagine anything, can create the impossible.",
    "We can only see a short distance ahead, but we can see plenty there that needs to be done.",
    "If a machine is expected to be infallible, it cannot also be intelligent.",
]

aug = naw.ContextualWordEmbsAug(model_path="bert-base-uncased", action="insert")  # BERT 모델을 사용하여 문맥에 어울리는 단어를 삽입(insert)하는 증강 객체를 생성합니다.
augmented_texts = aug.augment(texts)  # 텍스트 리스트에 대해 증강을 수행합니다.

for text, augmented in zip(texts, augmented_texts):  # 원본과 증강본을 비교하며 출력합니다.
    print(f"src : {text}")
    print(f"dst : {augmented}")
    print("------------------")
```

```python
import nlpaug.augmenter.char as nac  # 문자 단위 증강 도구를 임포트합니다.


texts = [  # 원본 텍스트 리스트
    "Those who can imagine anything, can create the impossible.",
    "We can only see a short distance ahead, but we can see plenty there that needs to be done.",
    "If a machine is expected to be infallible, it cannot also be intelligent.",
]

aug = nac.RandomCharAug(action="delete")  # 무작위로 문자를 삭제(delete)하는 증강 객체를 생성합니다.
augmented_texts = aug.augment(texts)  # 증강 수행

for text, augmented in zip(texts, augmented_texts):  # 결과 출력
    print(f"src : {text}")
    print(f"dst : {augmented}")
    print("------------------")
```

```python
import nlpaug.augmenter.word as naw  # 단어 단위 증강 도구


texts = [
    "Those who can imagine anything, can create the impossible.",
    "We can only see a short distance ahead, but we can see plenty there that needs to be done.",
    "If a machine is expected to be infallible, it cannot also be intelligent.",
]

aug = naw.RandomWordAug(action="swap")  # 무작위로 인접한 단어의 위치를 바꾸는(swap) 증강 객체 생성
augmented_texts = aug.augment(texts)

for text, augmented in zip(texts, augmented_texts):
    print(f"src : {text}")
    print(f"dst : {augmented}")
    print("------------------")
```

```python
import nlpaug.augmenter.word as naw  # 단어 단위 증강 도구


texts = [
    "Those who can imagine anything, can create the impossible.",
    "We can only see a short distance ahead, but we can see plenty there that needs to be done.",
    "If a machine is expected to be infallible, it cannot also be intelligent.",
]

aug = naw.SynonymAug(aug_src='wordnet')  # WordNet 유의어 사전을 사용하여 단어를 유의어로 교체하는 증강 객체 생성
augmented_texts = aug.augment(texts)

for text, augmented in zip(texts, augmented_texts):
    print(f"src : {text}")
    print(f"dst : {augmented}")
    print("------------------")
```

```python
import nlpaug.augmenter.word as naw


texts = [
    "Those who can imagine anything, can create the impossible.",
    "We can only see a short distance ahead, but we can see plenty there that needs to be done.",
    "If a machine is expected to be infallible, it cannot also be intelligent.",
]
reserved_tokens = [  # 증강에서 제외하거나 교환할 특별한 예약어 리스트 정의
    ["can", "can't", "cannot", "could"],
]

reserved_aug = naw.ReservedAug(reserved_tokens=reserved_tokens)  # 예약어 기반 증강(또는 보호) 객체 생성
augmented_texts = reserved_aug.augment(texts)

for text, augmented in zip(texts, augmented_texts):
    print(f"src : {text}")
    print(f"dst : {augmented}")
    print("------------------")
```

```python
import nlpaug.augmenter.word as naw


texts = [
    "Those who can imagine anything, can create the impossible.",
    "We can only see a short distance ahead, but we can see plenty there that needs to be done.",
    "If a machine is expected to be infallible, it cannot also be intelligent.",
]

back_translation = naw.BackTranslationAug(  # 역번역(Back Translation)을 통한 증강 객체 생성
    from_model_name='facebook/wmt19-en-de',   # 영어에서 독일어로 번역하는 모델
    to_model_name='facebook/wmt19-de-en'      # 독일어에서 다시 영어로 번역하는 모델
)
augmented_texts = back_translation.augment(texts)  # 역번역 증강 수행

for text, augmented in zip(texts, augmented_texts):
    print(f"src : {text}")
    print(f"dst : {augmented}")
    print("------------------")
```
