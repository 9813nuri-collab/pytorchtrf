```python
import torch  # 파이토치 라이브러리를 임포트합니다.
from torch import nn  # 신경망 관련 모듈(nn)을 임포트합니다.


x = torch.FloatTensor(  # 입력 데이터로 사용할 실수형 텐서를 생성합니다.
    [
        [-0.6577, -0.5797, 0.6360],  # 1번 데이터 샘플 (특성 3개)
        [0.7392, 0.2145, 1.523],    # 2번 데이터 샘플 (특성 3개)
        [0.2432, 0.5662, 0.322]      # 3번 데이터 샘플 (특성 3개)
    ]
)

print(nn.BatchNorm1d(3)(x))  # 3개의 특성에 대해 1차원 배치 정규화를 수행하고 결과를 출력합니다.
```
