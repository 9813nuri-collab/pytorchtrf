```python
for x, y in train_dataloader:  # 학습 데이터로더에서 배치 단위로 데이터를 가져옵니다.
    x = x.to(device)  # 입력 데이터를 설정된 장치(CPU/GPU)로 보냅니다.
    y = y.to(device)  # 정답 데이터를 설정된 장치로 보냅니다.

    output = model(x)  # 모델에 데이터를 입력하여 예측값을 얻습니다.

    _lambda = 0.5  # L1 정칙화의 강도를 조절하는 하이퍼파라미터 람다값입니다.
    l1_loss = sum(p.abs().sum() for p in model.parameters())  # 모든 파라미터의 절대값의 합을 구합니다 (L1 손실).

    loss = criterion(output, y) + _lambda * l1_loss  # 기존 손실 함수에 L1 손실을 더하여 최종 손실을 구합니다.
```
