```python
from torch import nn  # 신경망 모듈을 임포트합니다.


class Net(nn.Module):  # nn.Module을 상속받아 신경망 클래스를 정의합니다.
    def __init__(self):  # 생성자 함수 정의
        super().__init__()  # 부모 클래스의 생성자를 호출합니다.
        self.layer = nn.Sequential(  # 여러 층을 순차적으로 연결하는 컨테이너를 생성합니다.
            nn.Linear(1, 2),  # 입력 1, 출력 2의 선형 층을 정의합니다.
            nn.Sigmoid()      # 시그모이드 활성화 함수를 정의합니다.
        )
        self.fc = nn.Linear(2, 1)  # 입력 2, 출력 1의 선형 층을 정의합니다.
        self.apply(self._init_weights)  # 모델의 모든 모듈에 대해 _init_weights 함수를 재귀적으로 적용합니다.

    def _init_weights(self, module):  # 각 모듈(층)을 인자로 받아 초기화하는 함수입니다.
        if isinstance(module, nn.Linear):  # 현재 모듈이 선형 층(Linear)인 경우에만 실행합니다.
            nn.init.xavier_uniform_(module.weight)  # 가중치를 Xavier Uniform 방식으로 초기화합니다.
            nn.init.constant_(module.bias, 0.01)      # 편향을 0.01 상수로 초기화합니다.
        print(f"Apply : {module}")  # 초기화가 적용되는 모듈의 정보를 출력합니다.


model = Net()  # 신경망 모델 인스턴스를 생성합니다.
```
