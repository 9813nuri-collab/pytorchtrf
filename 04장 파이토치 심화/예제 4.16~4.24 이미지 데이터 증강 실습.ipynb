```python
from PIL import Image  # 파이썬 이미지 처리 라이브러리(PIL)를 임포트합니다.
from torchvision import transforms  # 파이토치에서 이미지 변환을 제공하는 transforms 모듈을 임포트합니다.


transform = transforms.Compose(  # 여러 변환을 하나로 묶어주는 Compose 객체를 생성합니다.
    [
        transforms.Resize(size=(512, 512)),  # 이미지의 크기를 512x512로 조정(Resize)합니다.
        transforms.ToTensor()                # PIL 이미지를 파이토치 텐서(Tensor)로 변환하고 0~1 사이로 정규화합니다.
    ]
)

image = Image.open("../datasets/images/cat.jpg")  # 원본 고양이 이미지 파일을 불러옵니다.
transformed_image = transform(image)              # 정의한 변환을 이미지에 적용합니다.
print(transformed_image.shape)                     # 변환된 이미지의 형태(채널, 높이, 너비)를 출력합니다.
plt.imshow(transformed_image)                     # 변환된 이미지를 시각화합니다.
```

```python
from PIL import Image
from torchvision import transforms


transform = transforms.Compose(
    [
        transforms.ColorJitter(  # 이미지의 색상을 무작위로 변경합니다.
            brightness=0.3, contrast=0.3,  # 밝기와 대비 조절
            saturation=0.3, hue=0.3         # 채도와 색상 조절
        ),
        transforms.ToTensor(),  # 연산을 위해 텐서로 변환
        transforms.Normalize(  # 이미지 데이터를 정규화합니다 (주로 ImageNet 통계값 사용).
            mean = [0.485, 0.456, 0.406],  # 각 채널별 평균값
            std = [0.229, 0.224, 0.225]    # 각 채널별 표준편차
        ),
        transforms.ToPILImage()  # 시각화를 위해 다시 PIL 이미지 형태로 변환합니다.
    ]
)

image = Image.open("../datasets/images/cat.jpg")
transformed_image = transform(image)
plt.imshow(transformed_image)
```

```python
import numpy as np  # 넘파이 라이브러리 임포트
np.bool = np.bool_ # Deprecated 오류 방지 (넘파이 버전에 따른 호환성 조치)
from PIL import Image
from torchvision import transforms
from imgaug import augmenters as iaa  # imgaug 라이브러리의 증강 도구들을 임포트합니다.


class IaaTransforms:  # imgaug를 파이토치 transforms에서 사용할 수 있는 클래스로 정의합니다.
    def __init__(self):
        self.seq = iaa.Sequential([  # imgaug의 순차적 증강 목록을 정의합니다.
            iaa.SaltAndPepper(p=(0.03, 0.07)),  # 이미지에 무작위로 점(소금/후추 노이즈)을 추가합니다.
            iaa.Rain(speed=(0.3, 0.7))           # 이미지에 비가 오는 효과를 추가합니다.
        ])
    
    def __call__(self, images):   # 클래스 호출 시 실행될 연산 정의
        images = np.array(images)  # PIL 이미지를 넘파이 배열로 변환합니다.
        print(images.shape, images.dtype)  # 데이터의 형태와 타입을 확인용으로 출력합니다.
        augmented = self.seq.augment_image(images)  # imgaug 증강을 수행합니다.
        return Image.fromarray(augmented)  # 다시 PIL 이미지 객체로 변환하여 반환합니다.


transform = transforms.Compose([
    IaaTransforms()  # 우리가 정의한 커스텀 imgaug 변환을 추가합니다.
])

image = Image.open("../datasets/images/cat.jpg")
transformed_image = transform(image)
plt.imshow(transformed_image)
```

```python
import numpy as np
from PIL import Image
from torchvision import transforms


class Mixup:  # 두 이미지를 합성하는 Mixup 증강을 위한 커스텀 클래스입니다.
    def __init__(self, target, scale, alpha=0.5, beta=0.5):
        self.target = target  # 함께 섞을 대상(Target) 이미지
        self.scale = scale    # 이미지 크기 스케일
        self.alpha = alpha    # 원본 이미지의 합성 비율
        self.beta = beta      # 대상 이미지의 합성 비율

    def __call__(self, image):
        image = np.array(image)  # 원본 이미지를 배열로 변환
        target = self.target.resize(self.scale)  # 대상 이미지의 크기를 원본에 맞춰 조정합니다.
        target = np.array(target)  # 대상 이미지를 배열로 변환
        mix_image = image * self.alpha + target * self.beta  # 두 이미지 배열을 비율에 맞게 가중 합산합니다.
        return Image.fromarray(mix_image.astype(np.uint8))  # 합성된 결과를 정수형 배열로 변환 후 이미지로 리턴합니다.


transform = transforms.Compose(
    [
        transforms.Resize((512, 512)),  # 이미지 크기를 512x512로 조정
        Mixup(  # Mixup 증강 적용
            target=Image.open("../datasets/images/dog.jpg"),  # 고양이 이미지에 섞을 강아지 이미지를 불러옵니다.
            scale=(512, 512),  # 강아지 이미지도 동일한 크기로 설정
            alpha=0.5,         # 원본 0.5 비율
            beta=0.5           # 타겟 0.5 비율
        )
    ]
)

image = Image.open("../datasets/images/cat.jpg")
transformed_image = transform(image)
plt.imshow(transformed_image)
```
